```python
from networkx import nx
```

```python
g = nx.read_weighted_edgelist('cbdb_data.txt')
```

```python
# 计算各 components 的大小
[len(c) for c in sorted(nx.connected_components(g), key=len, reverse=True)]
```

```python
# 总共有多少个 components
len(sorted(nx.connected_components(g), key=len, reverse=True))
```

```python
# 平均 degree
g.number_of_edges() * 2.0 / g.number_of_nodes()
```

```python
# Assortativity coefficient 同配系数
nx.degree_assortativity_coefficient(g)
```

```python
# 找到最大的 component，并转化为 graph
gcc = sorted(nx.connected_components(g), key=len, reverse=True)
giant = g.subgraph(gcc[0])
```

```python
# 计算 graph 的 diameter
nx.diameter(giant)
```
